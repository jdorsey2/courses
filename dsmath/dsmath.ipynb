## VBS

X = set of people in a clinical trial

$$
S = \{ x \in X : \text{x has VBS}\}  
$$

$$
H = \{ x \in X : \text{x does not have VBS} \}
$$


$$
P = \{ x \in X : \text{x tests positive for VBS}\}  
\\
N = \{ x \in X : \text{x tests negative for VBS}\}  
$$
 
$$
S \cup H = X
\\
S \cap H = \emptyset
\\
P \cup N = X
\\
P \cap N = \emptyset
$$

* $S \cap P$: true positive
* $H \cap N$: true negative
* $S \cap N$: false negative
* $H \cap P$: false positive
* $\frac{|S\cap P|} {|S|}$: true positive rate (closer to 1 is better)
* $\frac{|H\cap P|} {|H|}$: false positive rate (closer to 0 is better)
* $\frac{|S\cap N|} {|S|}$: false negative rate
* $\frac{|H\cap N|} {|H|}$: true negative rate

* $|S| \over |X|$: proprtion of people in the study with VBS
* $\frac{|H|}{|X|}$: proprtion of people in the study w/o VBS

## Numbers - The Real Number Line

#### Numbers - Algebra With Inequalities

if $a = b$, then $a + c = b + c$;  
if $a = b$ and $c \ne 0$, then $a * c = b * c$;  
if $a < b$, then $a + c < b + c$;  
if $a < b$ and if $c > 0$, then $a * c < b * c$;  
if $a < b$ and if $c < 0$, then $a * c > b * c$;

### Sigma Notation - Mean and Variance

$$X = \{ x_1, ..., x_n \}$$

mean of x: 
$$\mu_x = \frac{1}{n}  \sum _{i=1} ^n x_i$$ 

variance of x:
$$\sigma _x ^2 = \frac{1}{n} [ \sum _{i=1} ^n (x_i - \mu_x)^2 ]$$

standfard divation of x:
$$\sigma _x = \sqrt{\frac{1}{n} [ \sum _{i=1} ^n (x_i - \mu_x)^2} ]$$

Mean centering. new dataset has same relationship but mean is 0.

variance: how spreadout data is.

### Cartesian Plane - Point-Slope Formula for Lines

Slop of $\overrightarrow{AB}$: $M = \frac{d-b}{c-a}$

rise/row

Point-Slope Formula for Lines:

if a line l has slope M and if $(x_0, y_0)$ is any point on the line, then l has the equation:

$$y-y_0 = M(x-x_0)$$

### Cartesian Plane: Slope-Intercept Formula for Lines

if a line l has slope M and and l hits the y-axis at the point $(0, b)$ then l has the equation:

$$y = Mx + b$$
