## VBS

X = set of people in a clinical trial

$$
S = \{ x \in X : \text{x has VBS}\}  
$$

$$
H = \{ x \in X : \text{x does not have VBS} \}
$$


$$
P = \{ x \in X : \text{x tests positive for VBS}\}  
\\
N = \{ x \in X : \text{x tests negative for VBS}\}  
$$
 
$$
S \cup H = X
\\
S \cap H = \emptyset
\\
P \cup N = X
\\
P \cap N = \emptyset
$$

* $S \cap P$: true positive
* $H \cap N$: true negative
* $S \cap N$: false negative
* $H \cap P$: false positive
* $\frac{|S\cap P|} {|S|}$: true positive rate (closer to 1 is better)
* $\frac{|H\cap P|} {|H|}$: false positive rate (closer to 0 is better)
* $\frac{|S\cap N|} {|S|}$: false negative rate
* $\frac{|H\cap N|} {|H|}$: true negative rate

* $|S| \over |X|$: proprtion of people in the study with VBS
* $\frac{|H|}{|X|}$: proprtion of people in the study w/o VBS

### Functions - Increasing and Decreasing Functions

let $f: \mathbb{R} \rightarrow \mathbb{R}$. 
$f$ is strictly increasing if:
whenever $a<b$, we have $f(a)< f(b)$;
$f$ is strictly decreasing if:
whenever $a<b$, we have $f(a)> f(b)$.

#### horizontal line test

The function is strictly increasing or strictly decreasing if
whenever you graph it every single horizontal line has to intersect exactly once. 

### Functions - Composition and Inverse

$(g \circ f)(x) = g(f(x))$

f and g are inverses of each other: $g = f^{-1}$

if the graph of f fails the horizontal line test,
then f has no inverse.
it means that the only invertible functions are
ones that are either strictly increasing or strictly decreasing. 

## Tangent Lines - Slope of a Graph at a Point

tangent line to graph of $f(x)$ at $x=A$.
$$ f^{'} (a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h} $$

gives rate of change of $f(x)$ at $x=A$.

Derivative function

$f(x) = x^2$

$$
f^{'} (a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
\\
= \lim_{h \to 0} \frac{(a+h)^2 - a^2}{h}
\\
= \lim_{h \to 0} \frac{a^2 + 2ah + h^2 - a^2}{h}
\\
= \lim_{h \to 0} \frac{(2a+h)h}{h}
\\
= \lim_{h \to 0} (2a + h)
\\
= 2a
$$

## Bayes’ Theorem

1. product rule

$$P(A|B) = \frac{P(A, B)}{P(B)}$$

$$P(A|B) P(B) = P(A, B)$$

2. substitute P(B,A) for P(A,B) on right

$$P(A|B) P(B) = P(B,A)$$

3. apply product rule to the right side

$$P(A|B) P(B) = P(B|A) P(A)$$

4. get 

$$P(A|B) = \frac{P(B|A) P(A)}{P(B)}$$

inverse probability problem

Inverse probability problems are those where
the answer is in the form of the probability that a certain process with a certain probability
parameter is being used to generate the observed data. 


$B$: observed data

$A_i$: possible process i with probability parameter $\Theta_i$

### Example

Problem: let's say that we have two urns.
And one has 20% white marbles and one has 10% white marbles.
We observed three white marbles in a row being drawn with
replacement from one or other of the urns.
But we don't know which urn we are observing.
What is the probability that we are observing Urn 1 and
what is the probability that we are observing Urn 2? 


$$P(A_i|B) = \frac{P(B|A_i) P(A_i)}{P(B)}$$

using sum rule: $P(B) = P(\text{observed data} | parameter1)P(parameter1) + P(\text{observed data} | parameter2)P(parameter2) + ... + P(\text{observed data} | parametern)P(parametern)$


$$
P(A_1|B) = \frac{P(B|A_1) P(A_1)}{P(B, A_1) + P(B, A_2)}
\\
= \frac{P(B|A_1) P(A_1)}{P(B|A_1)P(A_1) + P(B|A_2)P(A_2)}
$$

principle of indifference: $P(A_1) = p(A_2) = .5$

Likelyhoods:

Urn1:
$$P(\text{3 white marbles in a row} | \text{20% white}) = .2 * .2 * .2 = 8/1000$$

Urn2:
$$P(\text{3 white marbles in a row} | \text{10% white}) = .1 * .1 * .1 = 1/1000$$

Sollution:

$$
P(A_1|B) = \frac{P(B|A_1) P(A_1)}{P(B|A_1)P(A_1) + P(B|A_2)P(A_2)}
\\
= \frac{8/1000 * 1/2}{8/1000 * 1/2 + 1/1000 * 1/2}
\\
= 8/9 = 88.9\%
$$

### updating probabilities based on new data

let's suppose that we have a fourth marble.
And it is also white.
How would we update our probabilities? 

$$
P(urn1) = P(A_1|B) = \frac{P(B|A_1) P(A_1)}{P(B|A_1)P(A_1) + P(B|A_2)P(A_2)}
\\
= \frac{.2 * 8/9}{ .2 * 8/9 + .1 * 1/9}
\\
= 16/17 = 94.12\%
$$

$$P(urn2) = 5.88\%$$

### Bayesian Inverse Probability

$$P(\Theta_i|D) = \frac{P(D|\Theta_i)P(\Theta_i)}{P(D)}$$

* $P(\Theta_i|D)$: posterior probability (after new data observed)
* $P(\Theta_i)$: prior probability (before any/new data observed)
* $P(D|\Theta_i)$: likelihood (standard/forward prob.)
* $P(D)$: marginal probability (of the data)
